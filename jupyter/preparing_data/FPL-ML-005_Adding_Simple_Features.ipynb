# FPL-ML-005: Adding Simple Features

## Import packages

```python
import pandas as pd
import numpy as np
```

## Read in data

```python
player_performance = pd.read_csv('../../data/csv/player_performance.csv')
player_performance.head()
```

```python
fixtures = pd.read_csv('../../data/csv/fixtures.csv')
fixtures.head()
```

```python
players = pd.read_csv('../../data/csv/players.csv')
players.head()
```

## Get additional player information

```python
# Number or rows in player_performance table
nrows = len(player_performance)

# Empty team array
team = np.zeros(nrows, dtype='int')
position = np.zeros(nrows, dtype='int')

# Loop through rows
for i in range(nrows):

    # Get team id
    current_fixture = player_performance.loc[i, 'fixture']
    if player_performance.loc[i, 'was_home']:
        team[i] = fixtures.loc[current_fixture, 'team_h']
    else:
        team[i] = fixtures.loc[current_fixture, 'team_a']

    # Get position
    current_id = player_performance.loc[i, 'id']
    position[i] = players.loc[current_id, 'position']

player_performance['team'] = pd.Series(team)
player_performance['position'] = pd.Series(position)
player_performance.head()
```

## Save table as csv

```python
player_performance.to_csv(r'../../data/csv/player_performance_01.csv',
                          index=False,
                          index_label=False)
```
